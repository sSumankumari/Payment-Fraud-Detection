### Data Processing

Now going to fit the model using the PCA
<b>Principal Component Analysis</b>: PCA is a most widely used tool in EDA and in machine learning for predictive models. Also an unsupervised statistical technique used to examine the interrelations among a set of variables.

```python
sc = StandardScaler()
scaled_df = sc.fit_transform(df[['average_rating', 'ratings_count', 'text_reviews_count']])
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(pca_data)
pca_df
```

```python
X = pca_df
distortions = []
for k in range(2,30):
    # Here "n_init=10" specifies that the k-means algorithm will be run 10 times with different centroid seeds and the final results will be the best output of the runs in terms of inertia.
    k_means = KMeans(n_clusters = k, n_init=10)
    k_means.fit(X)
    distortions.append(k_means.inertia_)
```

```python
fig = plt.figure(figsize=(10,6))
plt.plot(range(2,30), distortions, 'bx-')
plt.title("Elbow Curve")
```

##### Now using K-Means Clustering

```python
pca_df.columns = pca_df.columns.astype(str)

centroids = 7
k_means = KMeans(n_clusters=centroids, n_init=10)

pca_df['idx'] = k_means.fit_predict(pca_df)
pca_df
```

```python
#adding a new col 'rating_between' for future dummy
for i in range(5):
    df4.loc[ (df4['average_rating'] >= i) & (df4['average_rating'] <= i+1), 'rating_between'] = f"between {i} and {i+1}"
```

```python
#making a dummy col for each rating level
dummy_rating_df = df4['rating_between'].str.get_dummies(sep=",")
dummy_rating_df.head(3)
```

```python
#making a dummy col for each language
dummy_language_df = df4['language_code'].str.get_dummies(sep=",")
dummy_language_df.head()
```

```python
def weighted_rating(df):
    v = df['ratings_count']
    R = df['average_rating']
    C = df['average_rating'].mean()
    m = df['ratings_count'].quantile(0.9)
    
    return (v/(v+m) * R) + (m/(m+v) *C)
```

```python
df4['weighted_score'] = weighted_rating(df4)
```

```python
df4.columns
```

```python
df4.rename(columns={'0':'PC1', '1':'PC2', '2':'PC3'}, inplace=True)
df4.columns
```

## Training ML model

<b>Making a Baseline model for the dataframe</b>

```python
model_df = pd.concat([dummy_rating_df, dummy_language_df, df4['weighted_score'], df4['average_rating'], df3['ratings_count'], df4['PC1'], df4['PC2'], df4['PC3']], axis=1)
model_df
```

```python
min_max_scaler = MinMaxScaler()
scaled_model_df = min_max_scaler.fit_transform(model_df)
```

### 1. KNN Classification Algorithm
K-Nearest Neighbors (KNN) is one of the simplest algorithms used in Machine Learning for regression and classification problem. KNN algorithms use data and classify new data points based on similarity measures (e.g. distance function). ... The data is assigned to the class which has the nearest neighbors.

```python
knn_model = neighbors.NearestNeighbors(n_neighbors=centroids, algorithm='kd_tree')
knn_model.fit(scaled_model_df)
distances, indices = knn_model.kneighbors(scaled_model_df)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

### 2. DBSCAN
DBSCAN is a clustering algorithm that defines clusters as continuous regions of high density and works well if all the clusters are dense enough and well separated by low-density regions.

```python
dbscan_model = DBSCAN(eps=0.2, min_samples=4, algorithm='kd_tree')
dbscan_model.fit(scaled_model_df)
```

```python
scaled_model_df
```

```python
clusters = dbscan_model.labels_
np.unique(clusters)
```

```python
colors = ['royalblue', 'maroon', 'forestgreen', 'mediumorchid', 'tan', 'deeppink', 
          'olive', 'goldenrod', 'lightcyan', 'navy', 'red', 'orange', 'violet',
           'green', 'yellow', 'cyan', 'magenta', 'black', 'white', 'blue', 'steelblue']
vectorizer = np.vectorize(lambda x: colors[x % len(colors)])

plt.scatter(scaled_model_df[:,-1], scaled_model_df[:,-2], c=vectorizer(clusters))
```
